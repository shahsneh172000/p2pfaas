```python
# Number of the nodes
n_states = 3
lam = 2
mi = 1

# Transition rate probabilities matrix
Q_v = []

for i in range(n_states):
    Q_row = [0.0 for j in range(n_states)]
    
    new_mi = mi/i if i > 0 else 0
    new_lam = lam if i != n_states - 1 else 0
    
    Q_row[i] = (-1)*((new_mi) + new_lam)
    
    if i > 0:
        Q_row[i - 1] = new_mi
    
    if i != n_states - 1:
        Q_row[i + 1] = new_lam
        
    Q_v.append(Q_row)
    
Q = matrix(Q_v)

print(Q.determinant())
print(str(Q))

# Vector of evenly distributed probabilities
#p_v = [1.0 for i in range(n_states)]
#p = vector(p_v)

p = matrix.identity(n_states)

print(str(p))

for i in range(50):
    p = p*Q

print("\nOutput p = ")
print(str(p))
```

```python
def computeAlfa(l, m, states):
    # Number of the nodes
    n_states = states
    lam = l
    mi = m

    # vectors
    l_v = [lam for i in range(n_states)]
    m_v = [mi/i for i in range(1, n_states + 1)]

    #print("l_v = " + str(l_v))
    #print("m_v = " + str(m_v))

    # build the r vector
    r_v = [0 for i in range(n_states)]
    r_v[0] = 1
    for i in range(1, n_states):
        num = 1.0
        den = 1.0
        for j in range(i):
            num *= l_v[j]
            den *= m_v[j]
        r_v[i] = num/den

    r = vector(r_v)
    #print("r = " + str(r))

    # alfa
    alfa_v = [1/n_states for i in range(n_states)]
    r_sum = 0.0

    for i in range(n_states):
        r_sum += r_v[i]

    for i in range(n_states):
        alfa_v[i] = r_v[i]/r_sum

    alfa = vector(alfa_v)

    #print("alfa = " + str(alfa))
    return alfa

n_states = 10
for i in range(1, 40):
    alfa = computeAlfa(i, 1, 10)
    #print("#" + str(i) + " - " + str(alfa))
    out = ""
    for j in range(n_states):
        out += str(alfa[j]).replace(".", ",")
        out += " "
    print(out)
```
